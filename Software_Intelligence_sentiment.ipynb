Passo 5: Exibindo o resultado da primeira análise (um range entre -1 [avaliação negativa] e 1 [avaliação positiva])


```python
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob

# Carregar o modelo de linguagem
nlp = spacy.load('en_core_web_sm')

# Adicionar o pipe do spacytextblob
nlp.add_pipe('spacytextblob')

# Registar a extensão de polaridade
from spacy.tokens import Doc
if not Doc.has_extension('polarity'):
    Doc.set_extension('polarity', getter=lambda doc: doc._.blob.polarity)

# Texto de exemplo
user_input = 'This is a wonderful campsite. I loved the serenity and the birds chirping in the morning.'

# Processando o texto
doc = nlp(user_input)

# Acessando a polaridade
input_polarity = doc._.polarity

# Exibindo o resultado
sentiment = {'score': input_polarity}
print(sentiment)


```
